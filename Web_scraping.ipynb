```python
# # # scroll to bottom repeatedly to load more
# for _ in range(10):
#     driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
#     time.sleep(2)

SCROLL_PAUSE_TIME = 2
prev_count = 0

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)

    cards = driver.find_elements(By.CSS_SELECTOR, "div.mb-srp__card")
    curr_count = len(cards)

    print("Listings loaded:", curr_count)

    if curr_count>1100:
        break

    prev_count = curr_count
```

```python
def safe_text(parent, selector):
    elems = parent.find_elements(By.CSS_SELECTOR, selector)
    return elems[0].text if elems else None

# summary_list = house.find_elements(By.CSS_SELECTOR,"div.mb-srp__card__summary__list--item")
# for i in summary_list:
#     key = i.get_attribute('data-summary')
#     label = safe_text(i,"div.mb-srp__card__summary--label")
#     value = safe_text(i,"div.mb-srp__card__summary--value")
#     print(key,"->", label,value)
```

```python
houses = driver.find_elements(By.CSS_SELECTOR, "div.mb-srp__card")
final_data = []
for house in houses:
    title =safe_text(house, "h2.mb-srp__card--title")
    society =safe_text(house, "a.mb-srp__card__society--name")
    price =house.find_element(By.CSS_SELECTOR,"div.mb-srp__card__price--amount").get_attribute("textContent").strip()
    price_per_sqft =safe_text(house, "div.mb-srp__card__price--size")
    description =safe_text(house, "div.mb-srp__card--desc--text p")

    json_ld_element = house.find_element(
        By.CSS_SELECTOR,
        'script[type="application/ld+json"]'
    )
    json_text = json_ld_element.get_attribute("innerHTML")
    data = json.loads(json_text)
    no_of_rooms = data.get("numberOfRooms")
    type_of_residence = data.get('@type')
    site_url = data.get("url")
    latitude = data.get("geo", {}).get("latitude")
    longitude = data.get("geo", {}).get("longitude")
    seller_name = data.get("potentialAction", {}).get("seller", {}).get("name")
    locality = data.get('address',{}).get('addressLocality')
    region = data.get('address',{}).get('addressRegion')
    Country = data.get('address',{}).get('addressCountry')
    
    final_data.append({
        "Title":title,
        "Society":society,
        "Locality":locality,
        "Region":region,
        "Country":Country,
        "Total Price":price,
        "Price per Sqft":price_per_sqft,
        "latitude":latitude,
        "longitude":longitude,
        "No of Rooms":no_of_rooms,
        "Residence Type":type_of_residence,
        "Seller Name":seller_name,
        "Description":description,
        "Site URL":site_url,
    })

```

```python
import csv

def add_header_csv(filename, header_list):
    # Read the existing content
    with open(filename, 'r', newline='') as infile:
        reader = csv.reader(infile)
        data = list(reader) # Store all rows in a list

    # Add the header list as the first item in the data list
    data.insert(0, header_list)

    # Overwrite the file with the new data including the header
    with open(filename, 'w', newline='') as outfile:
        writer = csv.writer(outfile)
        writer.writerows(data) 
```

```python
add_header_csv('magicbricks_hyderabad.csv',header_list)
```
